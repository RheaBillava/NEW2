```python
import numpy as np
import matplotlib.pyplot as plt

#### CASE 1:  fs = 40*f;    delta = (2*pi*A*f)/fs = 0.785398

A = 5
f = 100
T = 2 / f
fs = 40 * f
t = np.arange(0, T, 1 / fs)
x = A * np.sin(2 * np.pi * f * t)

# Plotting the original sine wave
plt.figure(figsize=(20, 4))
plt.plot(t, x)
plt.title('Original Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Delta modulation
delta = 0.78539
xn_mod = np.zeros(len(x) + 1)  # Extend xn_mod by one element
d = np.zeros(len(x), dtype=int)

for i in range(len(x)):
    if x[i] > xn_mod[i]:
        d[i] = 1
        xn_mod[i + 1] = xn_mod[i] + delta
    else:
        d[i] = 0
        xn_mod[i + 1] = xn_mod[i] - delta
        
# Plotting both signals on the same graph
plt.figure(figsize=(20, 6))

# Plot the original sinusoidal signal
plt.plot(t, x, label='Original Signal')

# Plot the delta modulated signal
plt.step(np.arange(len(xn_mod) - 1) * T / len(x), xn_mod[:-1], label='Delta Modulated Signal')

plt.title('Comparison of Original Signal and Delta Modulated Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

# Encoding signal
plt.figure(figsize=(20, 6))
plt.step(np.arange(len(d)) * T / len(x), d, label='Encoded Signal', linestyle='--')
plt.title('Encoded Signal')
plt.xlabel('Time (s)')
plt.ylabel('Encoded Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting delta modulated signal
plt.figure(figsize=(20, 6))
plt.step(np.arange(len(xn_mod) - 1) * T / len(x), xn_mod[:-1], label='Decoded Signal')
plt.title('Decoded Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

# Demodulation
demodulated_signal = np.cumsum(xn_mod[1:])  # Demodulation
t_demod = np.linspace(0, T, len(demodulated_signal))  # Adjusting time scale

# Plotting the demodulated signal
plt.figure(figsize=(20, 4))
plt.plot(t_demod, demodulated_signal)
plt.title('Reconstructed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

In this experiment, we performed Delta Modulation (DM) for signal
transmission and reconstruction. Initially, a sinusoidal signal with
amplitude 5 and frequency 100 Hz was generated and plotted for
two cycles. The staircase approximation to this signal was then
created to represent the DM output.
Granular noise occurs with large delta values, resulting in larger
steps in the staircase approximation. This can cause the modulation
to overshoot or undershoot slow changes in the original signal
smaller deltas reduce granular noise but heightens the risk of slope
overloading, while larger deltas alleviate slope overloading but
introduce more granular noise

Quantization Effects: Delta modulation introduced quantization distortion,
approximating the sinusoidal waveform with a staircase-like pattern. Higher
quantization levels intensified distortion.
1.

Signal-to-Noise Ratio (SNR): Increasing step sizes in delta modulation
degraded SNR due to quantization noise, especially noticeable with larger
step sizes.
2.

Frequency Response: Delta modulation exhibited limitations in
reproducing higher frequency components due to finite sampling rates and
bandwidth constraints.
3.

Error Accumulation: Over iterations, error accumulation caused the
reconstructed signal to diverge gradually from the original sinusoid,
particularly in regions of rapid signal changes.
4.

Adaptation to Amplitude Changes: Delta modulation dynamically adjusted
to changes in signal amplitude, thanks to variable step sizes.
5.

Comparison with PCM: Delta modulation offered simplicity and reduced
bandwidth but was inferior to PCM in signal fidelity and noise immunity.
